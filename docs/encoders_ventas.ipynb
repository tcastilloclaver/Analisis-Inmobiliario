```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import os
import pandas as pd
import pickle

# Directorio para guardar los archivos pickle
pickle_dir = 'venta_pickles'
os.makedirs(pickle_dir, exist_ok=True)

# Cargar el archivo CSV actualizado
df = pd.read_csv("inmuebles_venta_sin_nan.csv")

# Limpieza de la columna 'CP': convertir a entero después de eliminar valores no enteros
df['CP'] = pd.to_numeric(df['CP'], errors='coerce')  # Convierte a número y asigna NaN a valores no convertibles
df['CP'].fillna(0, inplace=True)                     # Opcional: reemplaza NaN por un valor, aquí se usa 0 como ejemplo
df['CP'] = df['CP'].astype(int)                       # Finalmente, convierte a entero

# Encoders para columnas categóricas
label_encoders = {}
one_hot_encoders = {}

# Columnas que requieren Label Encoding (ej: columnas ordinales)
label_cols = ['Antigüedad', 'Conservación', 'Tipo De Casa']  

# Columnas que requieren One-Hot Encoding (ej: columnas nominales)
one_hot_cols = ['Tipo De Operación', 'Planta']  

# Target Encoding para `CP` basado en el precio promedio
# Se calcula el precio mediano de cada área y se asigna a una nueva columna
df['CP_encoded'] = df.groupby('CP')['Precio'].transform('median')

# Guardar el diccionario de target encoding para `CP` en un archivo pickle
target_encoding = df.groupby('CP')['Precio'].median().to_dict()
with open(os.path.join(pickle_dir, 'target_encoding_codigo_postal.pkl'), 'wb') as file:
    pickle.dump(target_encoding, file)

# Aplicar LabelEncoder y guardar en pickle
for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))  # Convertimos a str por posibles NaN
    label_encoders[col] = le
    # Guardar el LabelEncoder en un archivo pickle
    with open(os.path.join(pickle_dir, f'label_encoder_{col}.pkl'), 'wb') as file:
        pickle.dump(le, file)

# Aplicar One-Hot Encoding para 'Planta' sin agrupar las plantas superiores
for col in one_hot_cols:
    # Aplicar OneHotEncoder directamente sin agrupar
    ohe = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
    transformed_data = ohe.fit_transform(df[[col]].astype(str))
    one_hot_df = pd.DataFrame(transformed_data, columns=ohe.get_feature_names_out([col]))
    
    # Concatenar las columnas One-Hot con el DataFrame
    df = pd.concat([df, one_hot_df], axis=1)
    
    # Eliminar la columna original
    df = df.drop(columns=[col])
    
    one_hot_encoders[col] = ohe
    # Guardar el OneHotEncoder en un archivo pickle
    with open(os.path.join(pickle_dir, f'one_hot_encoder_{col}.pkl'), 'wb') as file:
        pickle.dump(ohe, file)

# Guardar el DataFrame procesado
df.to_csv('inmuebles_venta_encoding.csv', index=False)

print("Encodings aplicados y guardados. El archivo procesado está listo.")
```

Resumen de los Encoders Aplicados
1. LabelEncoder:
Objetivo: Transformar columnas categóricas con valores ordinales en números enteros únicos para cada categoría, manteniendo su orden.

Columnas aplicadas: Antigüedad, Conservación, y Tipo De Casa.

Funcionamiento:

LabelEncoder convierte cada categoría de estas columnas en un número entero único.
Es adecuado para variables categóricas ordinales, donde las categorías tienen un orden inherente (por ejemplo, Antigüedad podría tener categorías como Nueva, Usada, etc., que tienen un orden).
Guardado:

Cada LabelEncoder se guarda en un archivo .pkl en el directorio venta_pickles con el nombre label_encoder_{columna}.pkl.
2. OneHotEncoder:
Objetivo: Transformar columnas categóricas sin orden (nominales) en múltiples columnas binarias (0 o 1), una para cada categoría.

Columnas aplicadas: Tipo De Operación y Planta.

Funcionamiento:

Para Tipo De Operación: El One-Hot Encoding convierte cada categoría (por ejemplo, Venta o Alquiler) en una columna binaria que indica la presencia o ausencia de esa categoría.
Para Planta: Se aplica el One-Hot Encoding sin agrupar las plantas superiores, de modo que cada planta (1ª, 2ª, Bajo, etc.) tiene su propia columna binaria.
Este enfoque es útil para diferenciar cada planta como una categoría individual, sin reducir la granularidad de la información.
Guardado:

Cada OneHotEncoder se guarda en un archivo .pkl en el directorio venta_pickles con el nombre one_hot_encoder_{columna}.pkl.
3. Target Encoding para CP (Código Postal):
Objetivo: Asignar a cada código postal (CP) un valor basado en el precio promedio o mediano de las propiedades en esa área, para proporcionar un contexto geográfico en el análisis.

Funcionamiento:

Se calcula la mediana del precio en cada CP para evitar el efecto de valores atípicos.
Este valor de la mediana se asigna a una nueva columna, CP_encoded, que representa el precio típico de cada área.
Guardado:

El diccionario de Target Encoding para CP se guarda en un archivo .pkl en el directorio venta_pickles con el nombre target_encoding_codigo_postal.pkl.
