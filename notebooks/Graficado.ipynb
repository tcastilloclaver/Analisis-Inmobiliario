```python
import pandas as pd

# Load the data
file_path = 'venta_inmuebles_limpio.csv'
data = pd.read_csv(file_path)

# Display the first few rows and columns to understand the data structure
data.head(), data.columns
```

```python
import matplotlib.pyplot as plt



# 1. Histograma de la distribución del Precio
plt.hist(data['Precio'].dropna(), bins=30)
plt.title('Distribución del Precio')
plt.xlabel('Precio (en miles)')
plt.ylabel('Frecuencia')
plt.show()

# 2. Dispersión de Precio vs Superficie Construida
plt.scatter(data['Superficie Construida'], data['Precio'])
plt.title('Relación entre Precio y Superficie Construida')
plt.xlabel('Superficie Construida (m²)')
plt.ylabel('Precio (en miles)')
plt.show()

# 3. Gráfico de caja de Precio por Número de Habitaciones
data_box = data[['Precio', 'Habitaciones:']].dropna()
data_box = data_box[data_box['Habitaciones:'] <= 10]  # Limitar a <= 10 habitaciones para mejor visualización
data_box.boxplot(column='Precio', by='Habitaciones:', grid=False)
plt.title('Precio según Número de Habitaciones')
plt.suptitle('')
plt.xlabel('Número de Habitaciones')
plt.ylabel('Precio (en miles)')
plt.show()

# 4. Dispersión de Consumo Energético vs Emisiones CO2
plt.scatter(data['Consumo Energético'], data['Emisiones CO2'])
plt.title('Relación entre Consumo Energético y Emisiones CO2')
plt.xlabel('Consumo Energético')
plt.ylabel('Emisiones CO2')
plt.show()

# 5. Gráfico de barras de Precio según Tipo de Casa
data.groupby('Tipo de casa:')['Precio'].mean().plot(kind='bar')
plt.title('Precio Promedio según Tipo de Casa')
plt.xlabel('Tipo de Casa')
plt.ylabel('Precio Promedio (en miles)')
plt.show()

# 6. Gráfico de caja de Precio según Antigüedad
data_antig = data[['Precio', 'Antigüedad:']].dropna()
data_antig.boxplot(column='Precio', by='Antigüedad:', grid=False)
plt.title('Precio según Antigüedad del Inmueble')
plt.suptitle('')
plt.xlabel('Antigüedad')
plt.ylabel('Precio (en miles)')
plt.show()

# 7. Dispersión de Precio vs Superficie Útil
plt.scatter(data['Superficie útil:'], data['Precio'])
plt.title('Relación entre Precio y Superficie Útil')
plt.xlabel('Superficie Útil (m²)')
plt.ylabel('Precio (en miles)')
plt.show()
```
